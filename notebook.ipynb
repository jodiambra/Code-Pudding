```python
import pandas as pd 
import plotly_express as px
import plotly.graph_objects as go
```

### Full dataset

```python
# first dataset
url1 = 'https://media.githubusercontent.com/media/dataPracticum/codepudding/main/spotify/spotify_dataset.csv'
full = pd.read_csv(url1, on_bad_lines='skip', names=['user_id', 'artist_name', 'track_name', 'playlist_name'])
```

```python
# full dataset
full
```

```python
# info on the columns
full.info()
```

```python
# missing values
full.isna().sum()
```

```python
# top 100 most common artists
full.artist_name.value_counts().nlargest(100)
```

```python
# Top N artists 
px.bar(full.artist_name.value_counts().nlargest(100), title='Top 100 Most Common Artists')
```

```python
px.bar(full.artist_name.value_counts().nlargest(50), title='Top 50 Most Common Artists')
```

```python
# most common playlists
full.playlist_name.value_counts().nlargest(100)
```

```python
# percentage of values that are missing
full.artist_name.isna().sum() / len(full.artist_name) * 100
```

```python
# drop missing 
full.dropna(inplace=True)
```

```python
# Check size 
full.shape
```

```python
# overview 
full.describe()
```

```python
# unique artists
full.artist_name.nunique() / len(full.artist_name) * 100
```

```python
# most active users
full.user_id.value_counts().nlargest(20)
```

```python
full_user = full.groupby('user_id')[['artist_name', 'track_name', 'playlist_name']].value_counts(ascending=False)
full_user
```
