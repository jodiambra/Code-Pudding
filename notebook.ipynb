### Full dataset

```python
url1 = 'https://media.githubusercontent.com/media/dataPracticum/codepudding/main/spotify/spotify_dataset.csv'
full = pd.read_csv(url1, on_bad_lines='skip', names=['user_id', 'artist_name', 'track_name', 'playlist_name'])
```

```python
full
```

```python
full.info()
```

```python
# missing values
full.isna().sum()
```

```python
full.columns
```

```python
# percentage of values that are missing
full.artist_name.isna().sum() / len(full.artist_name) * 100
```

```python
# drop missing 
full.dropna(inplace=True)
```

```python
# Check size 
full.shape
```

```python
# overview 
full.describe()
```

```python
# unique artists
full.artist_name.nunique() / len(full.artist_name) * 100
```

```python
# most active users
full.user_id.value_counts().nlargest(20)
```

```python
full_user = full.groupby('user_id')[['artist_name', 'track_name', 'playlist_name']].value_counts(ascending=False)
full_user
```

### Top 10 dataset

```python
url2 = 'https://raw.githubusercontent.com/dataPracticum/codepudding/main/spotify/top_by_year/top10s.csv'
top10 = pd.read_csv(url2, encoding='latin-1')
```

```python
top10.columns
```

```python
top10.rename(columns={'top genre': 'genre'}, inplace=True)
```

```python
top10
```

```python
top10 = top10.drop('Unnamed: 0', axis=1)
```

```python
top10
```

```python
# summary stats
top10.describe()
```

```python
# Correlation matrix
top10.corr()
```

```python
top10.query()
```

```python
# most popular artists
top_artist = top10.artist.value_counts().nlargest(20)
```

```python
px.bar(top_artist, title='Top 20 Artists', height=900)
```

```python
# most popular genre
top_genre = top10.genre.value_counts().nlargest(20)
```

```python
px.bar(top_genre, title='Top 20 Genres', height=900)
```

```python
# most popular artists
top_year = top10.year.value_counts(sort=False)
```

```python
px.bar(top_year, title='Years With Most Popular Songs', height=900)
```

```python
top10
```

```python

px.scatter(top10, x='artist', y='pop', animation_frame='year', size='pop', animation_group='year', title='Top Artist per year', height=800)
```

### Unpopular dataset

```python
url3 = 'https://raw.githubusercontent.com/dataPracticum/codepudding/main/spotify/unpopular_songs.csv'
```

```python
unpopular = pd.read_csv('url3')
```

```python
unpopular
```

```python
unpopular.corr()
```

```python
unpopular.describe()
```

```python
unpopular.track_artist.value_counts().nlargest(20)
```

```python
artists = unpopular.groupby('track_artist')[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'explicit', 'popularity', 'track_name',
       'track_id']].value_counts()
display(artists)
```

```python
# Top 20 unpopular with highest danceability
unpopular.sort_values(by='danceability', ascending=False).nlargest(20, columns='danceability')
```

```python
# Top 20 unpopular with highest energy
unpopular.sort_values(by='energy', ascending=False).nlargest(20, columns='energy')
```

```python
popularity = unpopular.popularity.value_counts()
popularity
```

```python
px.histogram(unpopular.popularity, title='Popularity Ratings')
```
